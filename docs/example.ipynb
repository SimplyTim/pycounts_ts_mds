# Example usage

To use `pycounts_ts_mds` in a project:

```python
import pycounts_ts_mds

print(pycounts_ts_mds.__version__)
```

# Imports

```python
from pycounts_ts_mds.pycounts_ts_mds import count_words
from pycounts_ts_mds.plotting import plot_words
```

# Create a text file

We'll first create a text file to work with using a famous quote from Einstein:

```python
quote = """Insanity is doing the same thing
over and over and expecting different results."""
with open("einstein.txt", "w") as file:
    file.write(quote)
```

# Count words

We can count the words in our text file using the `count_words()` function. Note that this function removed punctuation and makes all words lowercase before counting.

```python
counts = count_words("einstein.txt")
print(counts)
```

# Plot words

We can now plot the result using hte `plot_words()` function:

```python
fig = plot_words(counts, n=5)
```
